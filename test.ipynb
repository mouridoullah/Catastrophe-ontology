```python
@app.route('/')
def page_ontologie():
    html = """<html><body>"""
    html += """<h2>Ontologie '%s' </h2>""" % onto.base_iri
    html += """<h3>Classes racines </h3>"""

    for Class in Thing.subclasses() :
        html += """<p><a href='%s'>%s</a></p>""" % (url_for('page_classe', iri = Class.iri), Class.name)

    html += """</html></body>"""

    return html
```

```python
@app.route('/classes/<path:iri>')
def page_classe(iri):
    Class = IRIS[iri]
    html = """<html><body><h2>Classe '%s' </h2>""" % Class.name
    html += """<h3>superclasses</h3>"""

    for SuperClass in Class.is_a:
        if isinstance(SuperClass, ThingClass):
            html += """<p><a href='%s'>%s</a></p>""" % (url_for('page_classe', iri = SuperClass.iri), SuperClass.name)
        else :
            html += """<p>%s</p>""" % SuperClass
    
    html += """<h3>Classes équivalentes</h3>"""
    for EquivClass in Class.equivalent_o:
        html += """<p>%s</p>""" % EquivClass

    html += """<h3>Sous-classes</h3>"""
    for SousClass in Class.subclasses():
        html += """<p><a href='%s'>%s</a></p>""" % (url_for('page_classe', iri = SousClass.iri), SousClass.name)
    
    html += """<h3>Individus</h3>"""
    for individu in Class.isinstance():
        html += """<p><a href='%s'>%s</a></p>""" % (url_for('page_individu', iri=individu.iri), individu.name)

    html += """</body></html>"""

    return html
```

```python
from rdflib import Graph

def recursive_import_search(graph, ontology_uri, processed_ontologies=None, imported_ontologies=None):
    if processed_ontologies is None:
        processed_ontologies = set()
    
    if imported_ontologies is None:
        imported_ontologies = set()
    
    # Si l'ontologie a déjà été traitée, retournez sans faire de traitement supplémentaire
    if ontology_uri in processed_ontologies:
        return
    
    # Ajoutez l'ontologie actuelle à la liste des ontologies traitées
    processed_ontologies.add(ontology_uri)
    
    # Chargez l'ontologie actuelle
    current_graph = Graph()
    current_graph.parse(ontology_uri, format="xml")
    
    # Recherchez les déclarations d'import dans l'ontologie actuelle
    import_triples = current_graph.triples((None, None, None))
    
    # Filtrez les triples pour obtenir les importations
    import_triples = [(str(s), str(o)) for s, p, o in import_triples if str(p) == 'http://www.w3.org/2002/07/owl#imports']
    
    # Imprimez les URI des ontologies importées dans l'ontologie actuelle
    for import_statement in import_triples:
        import_uri = import_statement[1]
        if import_uri not in imported_ontologies:
            print("Ontologie importée dans", ontology_uri, ":", import_uri)
            imported_ontologies.add(import_uri)
            
            # Appel récursif pour rechercher les importations dans l'ontologie importée
            recursive_import_search(graph, import_uri, processed_ontologies, imported_ontologies)

# Chargez l'ontologie A
g = Graph()
ontology_uri = "file8.rdf"
g.parse(ontology_uri, format="xml")

# Appelez la fonction récursive pour rechercher les importations
recursive_import_search(g, ontology_uri)
```

```python
def sévérité(kesaio, g, severite_niveau_0, severite_niveau_1, severite_niveau_2, seisme_uri, nombre_morts):
    if nombre_morts in severite_niveau_0:
        severite = "Niveau 0"
    elif nombre_morts in severite_niveau_1:
        severite = "Niveau 1"
    elif nombre_morts in severite_niveau_2:
        severite = "Niveau 2"
    else:
        severite = "Niveau inconnu"

        # Ajoutez une propriété de sévérité à l'instance de catastrophe
    g.add((seisme_uri, kesaio["sévérite_de_la_catastrophe"], Literal(severite, datatype=XSD.string)))
```
